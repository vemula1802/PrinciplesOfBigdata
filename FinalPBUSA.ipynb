```
def comp_func():
  diff_list = []
  k = " For the month "
  month_name = [ 'Jan','Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
  for x in range(0, len(my_list_2019)):
    diff_list.append(((my_list_2019[x] - my_list_2020[x])/(my_list_2019[x])) * 100)
    #print(diff_list[x])
    if diff_list[x] > 0:
      if diff_list[x] < 5:
        print("no clear conclusion" + k + month_name[x]+'.There was '+ str(round(diff_list[x])) +'% decrease in AQI')
      elif diff_list[x] < 10 :
        #more that 0-10% decrease in AQI
        print("Followed lockdown but not upto the mark"+ k + month_name[x]+'.There was '+ str(round(diff_list[x])) +'% decrease in AQI')
      elif diff_list[x] < 20:
         #more that 10-20% decrease in AQI
        print("Lockdown followed strictly"+ k + month_name[x]+'.There was '+ str(round(diff_list[x])) +'% decrease in AQI')
      else:
        #more that 20% decrease in AQI
        print("Followed very strictly"+ k + month_name[x]+'.There was '+ str(round(abs(diff_list[x]))) + '% increase in AQI')
    if diff_list[x] < 0 :
      if diff_list[x] > -5:
        print("No clear conclusion" + k + month_name[x] +'.There was '+ str(round(abs(diff_list[x]))) + '% increase in AQI')
      elif diff_list[x] > -10:
         print("Lockdown not followed strictly"+ k + month_name[x]+'.There was '+ str(round(abs(diff_list[x]))) + '% increase in AQI')
      else:
         print("Not followed at all"+ k + month_name[x])    
  return diff_list
```

```
diff_list = comp_func()
```
